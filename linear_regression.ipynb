```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/Salary_Data.csv")
x= dataset.iloc[:,[0]].values
y= dataset.iloc[:,1].values
print(x)
print(y)
```

```python
#splittung training testing
#train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
#training >>>>>> algorithm
from sklearn.linear_model import LinearRegression

#class >>>>>>>>> object
regressor = LinearRegression()
regressor.fit(x_train,y_train)

#testing
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
#evaluation
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,y_pred))
```

```python
A=regressor.coef_   #Slope
b=regressor.intercept_  #intersection
print("y=",A,"X +",b)
```

```python
print("the salary of the employee is : " , regressor.predict([[10.5]]))
```

```python
#visualization training

import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train , regressor.predict(x_train),color='blue')
plt.title("the salary vs year of experience graph")
plt.xlabel("years of experience ")
plt.ylabel("salary")
plt.show()
```

```python
#visualization testing

import matplotlib.pyplot as plt
plt.scatter(x_test, y_test, color='red')
plt.plot(x_train , regressor.predict(x_train),color='blue')
plt.title("the salary vs year of experience graph test")
plt.xlabel("years of experience ")
plt.ylabel("salary")
plt.show()
```
